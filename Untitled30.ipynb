```python
## ==========================================================
# SALES PREDICTION USING ADVERTISING DATASET
# COMPLETE PROJECT
# ==========================================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ==========================================================
# STEP 1: LOAD DATASET
# ==========================================================

df = pd.read_csv("Advertising.csv")

print("="*60)
print("FIRST 5 ROWS")
print("="*60)
print(df.head())

# ==========================================================
# STEP 2: DATA CLEANING
# ==========================================================

print("\nDataset Shape:", df.shape)

# Remove unnecessary column
if "Unnamed: 0" in df.columns:
    df.drop("Unnamed: 0", axis=1, inplace=True)

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Values")
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\nCleaned Dataset Shape")
print(df.shape)

# ==========================================================
# STEP 3: DATA EXPLORATION
# ==========================================================

print("\nStatistical Summary")
print(df.describe())

# ==========================================================
# STEP 4: CORRELATION ANALYSIS
# ==========================================================

plt.figure(figsize=(8,6))
sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()

# ==========================================================
# STEP 5: VISUALIZATION
# ==========================================================

features = ["TV", "Radio", "Newspaper"]

for col in features:

    plt.figure(figsize=(6,4))

    sns.scatterplot(
        x=df[col],
        y=df["Sales"]
    )

    plt.title(f"{col} vs Sales")

    plt.show()

# ==========================================================
# STEP 6: FEATURE SELECTION
# ==========================================================

X = df[["TV","Radio","Newspaper"]]
y = df["Sales"]

# ==========================================================
# STEP 7: DATA TRANSFORMATION
# ==========================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("\nData Standardization Completed")

# ==========================================================
# STEP 8: TRAIN TEST SPLIT
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTraining Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)

# ==========================================================
# STEP 9: MODEL TRAINING
# ==========================================================

model = LinearRegression()

model.fit(X_train, y_train)

print("\nModel Training Completed")

# ==========================================================
# STEP 10: PREDICTION
# ==========================================================

y_pred = model.predict(X_test)

# ==========================================================
# STEP 11: MODEL EVALUATION
# ==========================================================

mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print("\n" + "="*60)
print("MODEL PERFORMANCE")
print("="*60)

print("MAE :", round(mae,3))
print("MSE :", round(mse,3))
print("RMSE:", round(rmse,3))
print("R² Score:", round(r2,3))

# ==========================================================
# STEP 12: ACTUAL VS PREDICTED
# ==========================================================

results = pd.DataFrame({
    "Actual Sales": y_test,
    "Predicted Sales": y_pred
})

print("\nSample Predictions")
print(results.head(10))

# ==========================================================
# STEP 13: ACTUAL VS PREDICTED GRAPH
# ==========================================================

plt.figure(figsize=(8,6))

plt.scatter(y_test, y_pred)

plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")

plt.title("Actual vs Predicted Sales")

plt.show()

# ==========================================================
# STEP 14: RESIDUAL ANALYSIS
# ==========================================================

residuals = y_test - y_pred

plt.figure(figsize=(8,6))

plt.scatter(y_pred, residuals)

plt.axhline(y=0)

plt.xlabel("Predicted Sales")

plt.ylabel("Residuals")

plt.title("Residual Analysis")

plt.show()

# ==========================================================
# STEP 15: FUTURE SALES PREDICTION
# ==========================================================

future_campaign = pd.DataFrame({
    "TV":[250],
    "Radio":[40],
    "Newspaper":[60]
})

future_scaled = scaler.transform(future_campaign)

future_sales = model.predict(future_scaled)

print("\nPredicted Future Sales:")
print(round(future_sales[0],2))

# ==========================================================
# STEP 16: FEATURE IMPORTANCE
# ==========================================================

importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.coef_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

print("\nFeature Importance")
print(importance)

plt.figure(figsize=(8,5))

sns.barplot(
    data=importance,
    x="Feature",
    y="Importance"
)

plt.title("Advertising Impact on Sales")

plt.show()

# ==========================================================
# STEP 17: MARKETING BUDGET SIMULATION
# ==========================================================

simulation = pd.DataFrame({
    "TV":[50,100,150,200,250,300],
    "Radio":[5,10,20,30,40,50],
    "Newspaper":[10,20,30,40,50,60]
})

simulation_scaled = scaler.transform(
    simulation
)

simulation["Predicted Sales"] = model.predict(
    simulation_scaled
)

print("\nMarketing Budget Simulation")
print(simulation)

# ==========================================================
# STEP 18: SALES FORECAST GRAPH
# ==========================================================

plt.figure(figsize=(10,5))

plt.plot(
    simulation["TV"],
    simulation["Predicted Sales"],
    marker='o'
)

plt.xlabel("TV Advertising Budget")

plt.ylabel("Predicted Sales")

plt.title("Sales Forecast Based on Advertising Spend")

plt.grid(True)

plt.show()

# ==========================================================
# STEP 19: ADVERTISING IMPACT ANALYSIS
# ==========================================================

print("\nAdvertising Impact Analysis")

best_feature = importance.iloc[0]["Feature"]

print(
    f"Most influential advertising channel: {best_feature}"
)

# ==========================================================
# STEP 20: BUSINESS INSIGHTS
# ==========================================================

print("\n" + "="*60)
print("BUSINESS INSIGHTS")
print("="*60)

print("""
1. The model explains approximately 90% of sales variation.

2. Advertising spend strongly influences sales performance.

3. Radio and TV advertisements contribute more to sales growth.

4. Newspaper advertising has lower impact compared to TV and Radio.

5. Increasing advertising budget generally increases predicted sales.

6. Marketing teams should prioritize high-performing channels.

7. The model can forecast future sales before launching campaigns.

8. Budget simulation helps businesses optimize advertising investments.

9. Data-driven advertising decisions can improve ROI.

10. The model can support future marketing strategy planning.
""")
```
